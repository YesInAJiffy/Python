```python
import matplotlib.pyplot as plt

# Sample data (replace with your actual data)

'''
categories = ["Have Channels", "1000 Subs", "10K Subs", "100K", "1Mil", "10 Mil", "50 Mil"]
total_value = 2600  # 2.6 Bn Adjust this value to represent the total sum of all percentages
percentages = [x / total_value * 100 for x in [113.9, 10.1, 2, .3211, 32300/1000000, 1100/1000000, 35/1000000]]  # Calculate percentages
'''

categories = ["All", ">1000 and <10K", ">10K and <100K", ">100K and <1Mil", ">1 Mil and <10Mil", ">10 Mil and <50Mil", ">50 Mil"]
total_value = 113.9  # 2.6 Bn Adjust this value to represent the total sum of all percentages
users = [ 113900000, 10100000, 2000000, 321100, 32300, 1100, 35]
percentages = [x / total_value * 100 for x in [ 113.9, 10.1, 2, .3211, 32300/1000000, 1100/1000000, 35/1000000]]  # Calculate percentages
i =0
for number in percentages:
  formatted_number = "{:.6f}".format(number)  # Prints 0.0000001234
  user_count = "{:.0f}".format(users[i])  # Prints 0.0000001234
  out_of_x_str = f"1 out of {100 / number:.1f}"
  print(f"Channel Owners {user_count:12} : Subscribers {categories[i]:18} : Percentage {formatted_number:12} : {out_of_x_str}")
  i = i + 1



# Calculate bubble sizes based on percentages (adjust scaling factor as needed)
bubble_sizes = [p * 1000 for p in percentages]  # Scale percentage to desired bubble size

# Create a bubble chart
plt.figure(figsize=(8, 6))
plt.scatter(categories, [1 for _ in categories], s=bubble_sizes, c='blue', alpha=0.7, edgecolors='black')  # Set y to 1 for category labels

# Add category labels and title
plt.xlabel("Categories")
plt.ylabel("Percentage")
plt.title("Percentage Distribution by Category")

# Add percentages as text labels to bubbles (optional)
#for i, (txt, size) in enumerate(zip(categories, percentages)):
#    plt.text(i, 1 + size/200, f"{txt} - {size:.1f}%", ha='center')  # Adjust vertical position for readability

plt.xticks(rotation=45)  # Rotate category labels for better readability (optional)
plt.yticks([])  # Hide unnecessary y-axis ticks

plt.grid(True, linestyle='--', alpha=0.6, which='both', axis='x')  # Adjust grid style (optional)
plt.tight_layout()
plt.show()



numbers = percentages


# Scaling factor for bubble size (adjust as needed)
scaling_factor = 50000

# Create a scatter plot with bubble sizes based on numbers
plt.figure(figsize=(12, 10))
#plt.scatter(range(len(numbers)), numbers, s=[n * scaling_factor for n in numbers], c='purple', alpha=0.7, edgecolors='black')
plt.scatter(categories, numbers, s=[n * scaling_factor for n in numbers], c='purple', alpha=0.7, edgecolors='black')

# Add labels and title
plt.xlabel("Index")  # Replace with your specific label if needed
plt.ylabel("Subscriber Count")
plt.title("Bubble Size Represents Number")

# Optional: Add data point labels (numbers) above the bubbles
#for i, num in enumerate(numbers):
#    plt.text(i, num + 5, str(num), ha='center')  # Adjust vertical offset for readability

plt.grid(True)
plt.show()



```
