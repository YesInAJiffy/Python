```python
import matplotlib.pyplot as plt

# Sample data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
percentages = [30, 20, 15, 35]  # Percentages for each category

# Calculate bubble sizes based on percentages
# You can adjust the scaling factor as needed to control the bubble sizes
sizes = [percentage * 100 for percentage in percentages]

# Create bubble chart
plt.figure(figsize=(10, 6))
plt.scatter(range(len(categories)), [1] * len(categories), s=sizes, alpha=0.5)

# Customize appearance
plt.xticks(range(len(categories)), categories)
plt.xlabel('Categories')
plt.ylabel('Bubble Size')
plt.title('Bubble Chart Showing Percentage of Different Categories (Bubble Size Indicates Percentage)')
plt.grid(True)

# Display the chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data (replace with your actual data)
categories = ["Clothing", "Electronics", "Food & Beverages", "Home & Garden", "Entertainment", "Travel & Leisure"]
total_value = 200  # Adjust this value to represent the total sum of all percentages
percentages = [x / total_value * 100 for x in [30, 50, 20, 65, 45, 10]]  # Calculate percentages

# Calculate bubble sizes based on percentages (adjust scaling factor as needed)
bubble_sizes = [p * 1000 for p in percentages]  # Scale percentage to desired bubble size

# Create a bubble chart
plt.figure(figsize=(8, 6))
plt.scatter(categories, [1 for _ in categories], s=bubble_sizes, c='blue', alpha=0.7, edgecolors='black')  # Set y to 1 for category labels

# Add category labels and title
plt.xlabel("Categories")
plt.ylabel("Percentage")
plt.title("Percentage Distribution by Category")

# Add percentages as text labels to bubbles (optional)
for i, (txt, size) in enumerate(zip(categories, percentages)):
    plt.text(i, 1 + size/200, f"{txt} - {size:.1f}%", ha='center')  # Adjust vertical position for readability

plt.xticks(rotation=45)  # Rotate category labels for better readability (optional)
plt.yticks([])  # Hide unnecessary y-axis ticks

plt.grid(True, linestyle='--', alpha=0.6, which='both', axis='x')  # Adjust grid style (optional)
plt.tight_layout()
plt.show()
```
