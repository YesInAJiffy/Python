**PLOTTING IN PYTHON**

```python
# prompt: Basic Matplotlib example

import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [11, 13, 12, 15, 14]
# the first parameter is horizontal axis and the the second is vertical axis
plt.plot(x, y)
plt.show()

plt.scatter(x, y)
plt.show()
```

```python
# prompt: example of plt.xscale('log')

plt.scatter(x, y)
plt.xscale('log')
plt.show()
```

```python
# prompt: histogram example

plt.hist(y)
plt.show()

plt.hist(y, bins = 2)
plt.show()
```

```python
# prompt: an example which explains why scatter plots are important

import matplotlib.pyplot as plt

# **SCATTER PLOTS**

x = [1, 2, 3, 4, 5]
y = [11, 13, 12, 15, 14]
plt.scatter(x, y)
plt.show()

# The scatter plot shows that there is a positive linear relationship between x and y. This means that as x increases, y also increases. The strength of the relationship can be determined by the slope of the line that fits the data.

# Scatter plots can be used to identify outliers. An outlier is a data point that is significantly different from the rest of the data. Outliers can be caused by errors in data collection or by unusual events.

# Scatter plots can be used to identify trends. A trend is a pattern in the data that shows how one variable changes over time. Trends can be used to make predictions about future values.

# Scatter plots are a powerful tool for visualizing data. They can be used to identify relationships, outliers, and trends. Scatter plots are used in a wide variety of fields, including science, business, and finance.
```

```python
# Additional Plot Options

year = [1950, 1960, 1970, 1980, 1990, 2000, 2010, 2020]
population = [2.52, 2.77, 3.03, 3.34, 3.65, 4.01, 4.38, 4.76]
plt.plot(year, population)
plt.xlabel("Year")
plt.ylabel("Population")
plt.title("Population Graph")
plt.show()

#to change the ylabel start and numbers.
plt.plot(year, population)
plt.yticks([0,1,2,3,4,5,6,7])
plt.xlabel("Year")
plt.ylabel("Population")
plt.title("Population Graph")
plt.show()

#to change the ylabel start and numbers ALSO the labeling.
plt.plot(year, population)
plt.yticks([0,1,2,3,4,5], ["0B", "1B", "2B", "3B", "4B", "5B"])
plt.xlabel("Year")
plt.ylabel("Population")
plt.title("Population Graph")
plt.show()
```
