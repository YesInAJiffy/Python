**PLOTTING IN PYTHON**

```python
# prompt: an example which explains why scatter plots are important

import matplotlib.pyplot as plt

# **SCATTER PLOTS**

x = [1, 2, 3, 4, 5]
y = [11, 13, 12, 15, 14]
plt.scatter(x, y)
plt.show()

# The scatter plot shows that there is a positive linear relationship between x and y. This means that as x increases, y also increases. The strength of the relationship can be determined by the slope of the line that fits the data.

# Scatter plots can be used to identify outliers. An outlier is a data point that is significantly different from the rest of the data. Outliers can be caused by errors in data collection or by unusual events.

# Scatter plots can be used to identify trends. A trend is a pattern in the data that shows how one variable changes over time. Trends can be used to make predictions about future values.

# Scatter plots are a powerful tool for visualizing data. They can be used to identify relationships, outliers, and trends. Scatter plots are used in a wide variety of fields, including science, business, and finance.
```

```python
# Additional Plot Options

year = [1950, 1960, 1970, 1980, 1990, 2000, 2010, 2020]
population = [2.52, 2.77, 3.03, 3.34, 3.65, 4.01, 4.38, 4.76]
plt.plot(year, population)
plt.xlabel("Year")
plt.ylabel("Population")
plt.title("Population Graph")
plt.show()

#to change the ylabel start and numbers.
plt.plot(year, population)
plt.yticks([0,1,2,3,4,5,6,7])
plt.xlabel("Year")
plt.ylabel("Population")
plt.title("Population Graph")
plt.show()

#to change the ylabel start and numbers ALSO the labeling.
plt.plot(year, population)
plt.yticks([0,1,2,3,4,5], ["0B", "1B", "2B", "3B", "4B", "5B"])
plt.xlabel("Year")
plt.ylabel("Population")
plt.title("Population Graph")
plt.show()
```

**SIGNIFICANCE OF DICTIONARIES**

```python
# Usage of list

year = [1950, 1960, 1970, 1980, 1990, 2000, 2010, 2020]
population = [2.52, 2.77, 3.03, 3.34, 3.65, 4.01, 4.38, 4.76]

#Suppose we wish to get the population of 1980
index = year.index(1980)
print(population[index])

#Usage of Dictionary
# You need to use a key value pair in dictionary.
popdict = {1950:2.52, 1960 : 2.77, 1970:3.03, 1980:3.34}
print(popdict[1980])
```

```python
# prompt: dictionary example in Python

# Create a dictionary
my_dict = {
    "name": "John",
    "age": 30,
    "city": "New York"
}

# Access a value by key
print(my_dict["name"])  # John

# Add a new key-value pair
my_dict["state"] = "New York"

# Update a value by key
my_dict["age"] = 31

# Delete a key-value pair
del my_dict["state"]

# Add new item in Dictionary
my_dict["address"] ="honolulu"

#another way to remove
my_dict.pop("address")

# Print the dictionary
print(my_dict)  # {'name': 'John', 'age': 31, 'city': 'New York'}
```

```python
# prompt: create a pandas dataframe from list

import pandas as pd
names = ['Alice', 'Bob', 'Carol', 'Dave']
ages = [20, 25, 30, 35]
data = {'name': names,
        'age': ages}
df = pd.DataFrame(data)
print(df)

#Adding index column
indexes = ["A", "B", "C", "D"]
df.index = indexes
print(df)
```

```python
#Index as inplace means that the same df is modified, if inplace is false a new df is returned
import pandas as pd

# Create a DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 35],
        'City': ['New York', 'San Francisco', 'Los Angeles']}
df = pd.DataFrame(data)

# Setting 'Name' as the index without using inplace
df.set_index('Name')
# df still has the default integer index
print (df)
# Setting 'Name' as the index using inplace
df.set_index('Name', inplace=True)
# df is modified in place with 'Name' as the index
print(df)

print("***********")
#Accessing ITEMS via indexes
row_alice = df.loc['Alice']
print(row_alice)

print("***********")
rows_bob_to_charlie = df.loc['Bob':'Charlie']
print(rows_bob_to_charlie)

print("***********")
#Revert back to the default integer index
df.reset_index(inplace=True)
print(df)

print("***********")
df.set_index('Age', inplace=True)
print(df)

print("***********")
young_people = df[df.index < 30]
print(young_people)
```

**LOC and iLOC in Pandas**

LOC is label based

iLOC is Position based

```python
# prompt: example of LOC in pandas with row and column selection

df = pd.DataFrame({'Name': ['Alice', 'Bob', 'Carol', 'Dave'],
                   'Age': [20, 25, 30, 35]})

df.set_index('Name', inplace=True)
print("***********")

# Select the row with the label 'Bob'
print(df.loc['Bob'])
print("***********")

# Select the column 'Age'
print(df['Age'])
print("***********")

# Select the row with the label 'Bob' and the column 'Age'
print(df.loc['Bob', 'Age'])
print("***********")

# Select the rows with the labels 'Bob' and 'Carol'
print(df.loc[['Bob', 'Carol']])
print("***********")

# Select the columns 'Name' and 'Age'
#print(df[['Name', 'Age']])

# Select the rows with the labels 'Bob' and 'Carol' and the columns  'Age'
print(df.loc[['Bob', 'Carol'], [ 'Age']])
print("***********")

# All rows and only Age Column
print(df.loc[:, ["Age"]])
```

```python
# iloc is based on the position

df = pd.DataFrame({'Name': ['Alice', 'Bob', 'Carol', 'Dave'],
                   'Age': [20, 25, 30, 35]})


# Select the row with the index 1
print(df.iloc[1])

# Select the column 'Age'
print(df.iloc[:, 1])

# Select the row with the index 1 and the column 'Age'
print(df.iloc[1, 1])

# Select the rows with the indices 1 and 2
print(df.iloc[1:3])

# Select the columns 'Name' and 'Age'
print(df.iloc[:, [0, 1]])

# Select the rows with the indices 1 and 2 and the columns 'Name' and 'Age'
print(df.iloc[1:3, [0, 1]])

# All rows and only Age Column
print(df.iloc[:, [1]])
```

**LOGICAL AND NUMPY**

```python
# prompt: example of logical And and Logical or on numpy array

import numpy as np
a = np.array([1, 2, 3, 4, 5])
b = np.array([2, 3, 4, 5, 6])

# Logical AND
print(np.logical_and(a > 3, b >3))

# Logical OR
print(np.logical_or(a<3, b<3))
```
