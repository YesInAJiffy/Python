**PLOTTING IN PYTHON**

```python
# prompt: Basic Matplotlib example

import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [11, 13, 12, 15, 14]
# the first parameter is horizontal axis and the the second is vertical axis
plt.plot(x, y)
plt.show()

plt.scatter(x, y)
plt.show()
```
